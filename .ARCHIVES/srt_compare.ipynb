```python
from lib_srt import srt
from tabulate import tabulate

folder = "tests/"
folder = "D:/Downloads/JD2/The Chosen (mq stfr)/srt"
srt_1 = "The.Chosen.S01E08.720p.BluRay.x265.10Bit-Pahe.li.srt"
srt_2 = "The.Chosen.S01E08.720p.BluRay.x265.10Bit-Pahe.li.fr.srt"

s1 = srt()
s2 = srt()

s1.load(folder+'/'+srt_1), s2.load(folder+'/'+srt_2)
```

```python
s1.all(), s2.all()
```

```python
h1 = s1.srt_to_html()
with open("test.html", "w", encoding="utf-8") as fout:
    fout.write(h1)

cont = True
cpt=0
res=[]
while cont and cpt < 5:
    sub1=s1.next()
    sub2=s2.next()
    if sub1=={} or sub2=={}:
        cont = False
    else:
        res.append([sub1['id'],sub1['text'],sub2['id'],sub2['text']])
    cpt += 1
print(tabulate(res))
```

```python
translated = ""
with open("test.html", "r") as fin:
    translated = fin.read()
srt_ready = s1.html_to_srt(translated)
print(srt_ready)
```
